# Splitting

```r
nrow(health_data[health_data$Target == 0,])*100/nrow(health_data)
nrow(health_data[health_data$Target == 1,])*100/nrow(health_data)
```

```r
# The function "createDataPartition" can be used to create balanced splits of the data.

train_rows <- createDataPartition(health_data$Target, p = 0.7, list = F)
train_data <- health_data[train_rows, ]
dim(train_data)
```

```r
val_data <- health_data[-train_rows, ]
dim(val_data)
```

```r
nrow(train_data[train_data$Target==0,])*100/nrow(train_data)
nrow(train_data[train_data$Target==1,])*100/nrow(train_data)                                            
nrow(val_data[val_data$Target==0,])*100/nrow(val_data)                                            
nrow(val_data[val_data$Target==1,])*100/nrow(val_data)
```

# Standardization

```r
std_model <- preProcess(train_data[, !names(train_data) %in% c("Target")], method = c("center", "scale"))
std_model

# method = "center" subtracts the mean of the predictor's data (from the data in x) from the predictor values
# method = "scale" divides by the standard deviation.

# The predict() function is used to standardize any other unseen data
```

```r
train_data[, !names(train_data) %in% c("Target")] <- predict(object = std_model, newdata = train_data[, !names(train_data) %in% c("Target")])
```

```r
val_data[, !names(val_data) %in% c("Target")] <- predict(object = std_model, newdata = val_data[, !names(val_data) %in% c("Target")])
```

```r
pc_traindata=cbind(myPr$x,Target=train_data[, names(train_data) %in% c("Target")])
```

```r
pc_traindata=data.frame(pc_traindata)
head(pc_traindata)
```

```r
dim(train_data)
```

```r
dim(pc_traindata)
```

```r
myPr_val <- prcomp(val_data[, !names(train_data) %in% c("Target")])
```

```r
pc_valdata=data.frame(cbind(myPr_val$x,Target=val_data[, names(val_data) %in% c("Target")]))
```

```r
model <- glm(formula = Target~PC1+PC2+PC3+PC4+PC5+PC6+PC7+PC8+PC9+PC10+PC11+PC12, data =pc_traindata)
val_pred=predict(model, pc_valdata[, !(names(pc_valdata) %in% c("PC13", "PC14", "PC15", "PC16", "PC17", "PC18", "PC19", "PC20", "PC21", "PC22", "Target"))])
```

```r
length(val_pred)
```

```r
length(val_data$Target)
```

```r
val_pred = ifelse(val_pred > 0.5, 1, 0)
```

```r
conf_matrix = table(val_data$Target, val_pred)
conf_matrix
```

```r
recall = conf_matrix[2,2]/sum(conf_matrix[2,])
recall

precision <- conf_matrix[2, 2]/sum(conf_matrix[,2])
precision
```
